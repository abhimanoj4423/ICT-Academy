## **Finding Outliers**

```python
plt.boxplot(iris[['Sepal Length','Sepal Width','Petal Length','Petal Width']])
```

## Relationship between **Sepal Length** and **Sepal Width** 

```python
colors = ['red','orange','blue']
j=0
groups = iris.groupby("Classification")
for key,i in groups:
    plt.scatter(i["Sepal Length"], i["Sepal Width"], c=colors[j], label=key,marker='^')
    j=j+1
plt.legend()
plt.title("Relationship between Sepal Length and Sepal Width",fontsize=14)
plt.xlabel("Sepal Length in cm")
plt.ylabel("Sepal Width in cm")
```

## Relationship between **Petal Length** and **Petal Width**

```python
colors = ['red','orange','blue']
j=0
groups = iris.groupby("Classification")
for key,i in groups:
    plt.scatter(i["Petal Length"], i["Petal Width"], c=colors[j], label=key,marker='^')
    j=j+1
plt.legend()
plt.title("Relationship between Petal Length and Petal Width",fontsize=14)
plt.xlabel("Petal Length in cm")
plt.ylabel("Petal Width in cm")
```

## Relationship between **Petal Length** and **Sepal Length**

```python
colors = ['red','orange','blue']
j=0
groups = iris.groupby("Classification")
for key,i in groups:
    plt.scatter(i["Petal Length"], i["Sepal Length"], c=colors[j], label=key,marker='^')
    j=j+1
plt.legend()
plt.title("Relationship between Petal Length and Sepal Length",fontsize=14)
plt.xlabel("Petal Length in cm")
plt.ylabel("Sepal Length in cm")
```
